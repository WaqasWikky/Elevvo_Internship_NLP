```python
from datasets import load_dataset
from transformers import AutoTokenizer, AutoModelForQuestionAnswering, pipeline
import string
import re
```

```python
# Load 5% of the training set and 5% of validation for quick testing
dataset = load_dataset("squad")
train_dataset = dataset['train'].select(range(int(0.05 * len(dataset['train']))))
val_dataset = dataset['validation'].select(range(int(0.05 * len(dataset['validation']))))

print("Sample QA Example:")
print("Context:", train_dataset[0]['context'])
print("Question:", train_dataset[0]['question'])
print("Answer:", train_dataset[0]['answers']['text'])
```

```python
model_name = "distilbert-base-uncased-distilled-squad"  # ~250 MB
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForQuestionAnswering.from_pretrained(model_name)
```

```python
qa_pipeline = pipeline("question-answering", model=model, tokenizer=tokenizer)
```

```python
context = """
The Apollo program was the third United States human spaceflight program carried out by NASA,
which accomplished landing the first humans on the Moon from 1969 to 1972. First conceived during
Eisenhower’s administration as a three-man spacecraft to follow the one-man Project Mercury which
put the first Americans in space, Apollo was later dedicated to President John F. Kennedy’s national goal
of “landing a man on the Moon and returning him safely to the Earth” by the end of the 1960s.
"""

question = "Who was the Apollo program dedicated to?"

result = qa_pipeline(question=question, context=context)
print("Answer:", result['answer'])
```
